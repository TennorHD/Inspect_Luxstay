```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
```

```python
driver = webdriver.Chrome(executable_path=r"C:\Users\Owner\Desktop\DS\chromedriver\chromedriver.exe")
```

```python
name=[] #List to store name of the product
category=[]
description=[]
prices=[] #List to store price of the product
location=[] #List to store rating of the product
```

```python
for i in range(200):
    driver.get("https://www.luxstay.com/s/?page={}".format(i+1))
    content = driver.page_source
    soup = BeautifulSoup(content)
    for a in soup.findAll('div', attrs={'class':'col-xs-6 col-md-3 col-lg-20'}):
        cat=a.find('div', attrs={'class':"promo__type bold"})
        des=a.find('div', attrs={'class':'promo__description my--6 text-lowercase'})
        tit=a.find('div', attrs={'class':'promo__title extra-bold'})
        pri=a.find('div', attrs={'class':'promo__price mb--6'})
        loc=a.find('div', attrs={'class':'promo__add mb--6 promo__add mb--6'})
        if cat == None:
            category.append('Invalid')
        else:
            category.append(cat.text)
        if tit == None:
            name.append('Invalid')
        else:
            name.append(tit.text)
        if des == None:
            description.append('Invalid')
        else:
            description.append(des.text)
        if pri == None:
            prices.append('Invalid')
        else:
            prices.append(pri.text)
        if loc == None:
            location.append('Invalid')
        else:
            location.append(loc.text)
```

```python
import re
name_temp = []
for i in range(len(name)):
    if name[i] != 'Invalid':
        try:
            name_temp.append(re.split(r'\n            |\n          ',name[i])[1])
        except:
            print(i)
            break
    else:
        name_temp.append(name[i])
df1 = pd.DataFrame(name_temp,columns = ['name'])
print(df1)
```

```python
cat_temp = []
for i in range(len(category)):
    if category[i] != 'Invalid':
        temp = re.split(r'\n        |\n         | ·  ',category[i])
        if len(temp)<2:
            cat_temp.append(('None','None'))
        elif len(temp)<4:
            cat_temp.append((temp[1],'None'))
        else:
            cat_temp.append((temp[1],temp[3]))
    else:
        cat_temp.append(('None','None'))
df2 = pd.DataFrame(cat_temp, columns = ['category','project'])
print(df2)
```

```python
des_temp = []
for i in range(len(description)):
    if description[i] != 'Invalid':
        temp = re.split(r' guests · | guest · | Bedroom · | Bedrooms · | Bedrooms| Bedroom| bathrooms| bathroom',description[i])[0:3]
        temp = tuple([x if x else '0' for x in temp])
        des_temp.append(temp)
    else:
        des_temp.append(('0','0','0'))
df3 = pd.DataFrame(des_temp, columns = ['guest','bedroom','bathroom'])
print(df3)
```

```python
pri_temp = []
for i in range(len(prices)):
    if prices[i] != 'Invalid':
        temp = re.split(r'/night \$|/night | \$',prices[i])
        if temp[2]:
            pri_temp.append((temp[1],'1',temp[2]))
        else:
            pri_temp.append((temp[1],'0','Invalid'))
    else:
        pri_temp.append(('0','0','Invalid'))
df4 = pd.DataFrame(pri_temp, columns = ['price','discounted','normal price'])
print(df4)
```

```python
loc_temp = [tuple(re.split(r', ',location[i])) if location[i] != 'Invalid' else ('Invalid','Invalid','Invalid') for i in range(len(location))]
df5 = pd.DataFrame(loc_temp, columns = ['province','state','country'])
print(df5)
```

```python
print(len(name_temp),len(cat_temp),len(des_temp),len(pri_temp),len(loc_temp))
```

```python
df = pd.concat([df1,df2,df3,df4,df5],axis = 1)
df.to_csv("./result200.csv",encoding='utf-8-sig')
df
```
